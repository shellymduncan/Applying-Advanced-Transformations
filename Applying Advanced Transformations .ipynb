# Load superhero info dataset

```python
# load data
info = pd.read_csv('Data/superhero_info - superhero_info.csv')
info.head()
```

```python
info.info()
```

```python
# check for missing data
info.isna().sum()
```

```python
# check for duplicates
info.duplicated().sum()
```

```python
# look at the Hero|Publisher column
info['Hero|Publisher'].head()
```

```python
# split the Hero|Publisher column into Hero(just the name of the Hero) and Publisher columns
info['Hero|Publisher'].str.split('|', expand = True)
```

```python
# save the 2 new columns into the dataframe
info[['Hero', 'Publisher']] = info['Hero|Publisher'].str.split('|', expand = True)
info.head()
```

```python
# drop the original colum Hero|Publisher
info = info.drop(columns = ['Hero|Publisher'])
info.head()
```

```python
# examin the measurements column
info['Measurements'].head()
```

```python
# split the measurements column into Height(numeric) and Weight(numeric) columns
# examining a single value from the coordinates column as a test
test_meas = info.loc[0, 'Measurements']
print(type(test_meas))
test_meas
```

```python
# replace single quotes with double quotes
test_meas = test_meas.replace("'", '"')
test_meas
```

```python
# apply json.loads
fixed_meas = json.loads(test_meas)
print(type(fixed_meas))
```

```python
# now apply to the entire column
info['Measurements'] = info['Measurements'].str.replace("'", '"')
# apply json.loads
info['Measurements'] = info['Measurements'].apply(json.loads)
```

```python
# check the column
info['Measurements']
```

```python
# check a single value after transformation
test_meas = info.loc[0, 'Measurements']
print(type(test_meas))
test_meas
```

```python
# use .apply pd.Series to convert a dict to columns
Height_Weight = info['Measurements'].apply(pd.Series)
Height_Weight
```

```python
# check the dataframe
info = pd.concat((info, Height_Weight), axis = 1)
info.head()
```

```python
# split the Height column to remove cm
info['Height'].str.split(' ', expand = True)
```

```python
info[['Height(cm)', 'cm']] = info['Height'].str.split(' ', expand = True)
info.head()
```

```python
# split the Weight column to remove kg
info['Weight'].str.split(' ', expand = True)
```

```python
info[['Weight(kg)', 'kg']] = info['Weight'].str.split(' ', expand = True)
info.head()
```

```python
# remove Measurements, cm and kg columns
info = info.drop(columns = ['Measurements', 'cm', 'kg', 'Height', 'Weight'])
info.head()
```

```python
# change Height(cm) and Weight(kg) to numeric(float)
info['Height(cm)'] = info['Height(cm)'].astype(float)
info['Weight(kg)'] = info['Weight(kg)'].astype(float)
info.info()
```

# Load superhero power dataset

```python
# load data
powers = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
powers.head()
```

```python
# check the dataypes
powers.info()
```

```python
# split the powers column 
powers['Powers_split'] = powers['Powers'].str.split(',')
powers.head()
```

```python
#exploding the columns
exploded = powers.explode('Powers_split')
exploded[['hero_names','Powers','Powers_split']].head()
```

```python
# saving the unique values from the exploded column
cols_to_make = exploded['Powers_split'].dropna().unique()
cols_to_make
```

```python
# create new columns using a for loop and .str.contains
for col in cols_to_make:
    powers[col] = powers['Powers'].str.contains(col)
powers.head()
```

```python
# merge both info and powers datasets
super_hero = pd.merge(info, powers, left_on = 'Hero', right_on = 'hero_names')
super_hero.head()
```

```python
# drop the old columns
super_hero = super_hero.drop(columns = ['hero_names', 'Powers', 'Powers_split'])
```

```python
# the final datset
super_hero.head()
```
